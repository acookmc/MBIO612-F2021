```r
###Question 4
vector_of_sites <- c("site_1", "site_2", "site_3", "site_4", "site_5", "site_6")
vector_of_abundance <- c(10, 21, 32, 6, 22, 5)
numerical_average_abundance = mean(vector_of_abundance)
df_site_abudance= rbind(vector_of_sites, vector_of_abundance)
site_max_abundace=df_site_abudance[1, which.max(df_site_abudance[2,])]
my_list=(list(vector_of_sites, vector_of_abundance, numerical_average_abundance, site_max_abundace))
names(my_list)= c("Vector of Sites", "Vector of Abundances", "Numerical Average Abundance Value", "The Site with Highest Abundance")
print(my_list)
```
