### Some definitions

- <i>Experiment</i> is a procedure, or random process that yields one outcome out of all the possible outcomes of the experiment. Some examples of experiments are:
  - Rolling a die 
  - Counting the number of empty slots in a Biki stand 
  - Measuring the weight of shark that was captured
  - Observing the number of bleacher corals at a given spot
<br/>  


#### Sample Space
- <i>Sample Space S</i> is the exhaustive set of all possible, disjoint outocmes of an experiment 
  - when rolling a die, its the finite set ${1, 2, 3, 4, 5, 6}$ 
  - When counting the number of empty slots in a Biki stand, it's a value greater or equal to 0 less than or equal to the total number of slots available
    - Discrete countably infinite set
  - When weighing patients in a hospital, it's the continuous infinite set [0.X, 1000] pounds


#### Sample Space - Cont'd
- When sampling from a uniform range [0,99], the sample space consists of all the integers in that range.
    
- Exploring the sample space for some common discrete experiment requires enumerating (or counting) all possibilities  
  - Many counting problems requires understanding and being able to compute __permutations__ and __combination__
  - In trivial cases, you can simulate the space of all possible events and subsequently compute the propbability of an event.
  
- Ex. What the probability of guessing someone's 4 digit PIN in the first try.
  - What is the sample space?
- What is the probability of  guessing someone's 4 digit PIN in the first try if they tell you that thier PIN does not contain any repeated digits
  - What is the sample space?

#### Event 
- In an experiment, an <i>avent E</i> represents a subset of the sample space 
  - The die lands 4
  - The die lands on a odd side (1, 3 or 5)
  - The number of empty Biki slots is 0 
  -  The weight of a patient is greater than 140 Kgs but less than 200 Kgs

#### Probability

- Probabilities quantifying beliefs. 
- The are used to measure uncertainty of an event. 
- Ex:
  - How likely is a shark to attach at Kaneohe today?
  - How unlikely is a given person to donate to a fundraising campaign?
  - How likely is a person to contract the flu this season?
  
- Intuitively, we talk of a `null` (0) probability when we know an event is not possible, and a probability of 1 if an event is certain.

- Formally, probability is a number that quantifies the likelihood of an event and which satisfies the following two properties:
 
 $~~~~~~~~~~~~~~0 <= p(s) <= 1$$~~~~~~~~~~~~\text{and}~~~~~~~~ \sum\limits_{s \in S}{p(s) = 1}$

A number that is between 0 a 1 is not necessarily a probability unless the sum of all outcomes in the sample space is 1.



### Long run frequencies of event

- The concept of long-term is an essential aspect of probability 
  - In a casino game where your chances of winning are 1 in 4, you are not guaranteed to win if you participate four times
  

- When flipping a coin -- assuming the coin is unbiased -- we would intuitively expect to observe `in the long run` as many heads as tails, or about 50% heads and 50% tails.
  - The probabilities of heads and tails are $p(H)=0.5$ and $p(T)=0.5$ respectively 
  
  
- We can prove this computationally by generating heads (ones) and tails (zeros) using the `Python`.

```python
from scipy.stats import bernoulli

randomTrials = bernoulli(0.99).rvs(100)

print(randomTrials[0:20])
```

- The progression of the number of heads after each iteration can be obtained using the cumulative sum (or `cumsum`). 

- The cumulative sum for an element at position $a$, where $a > 1$ of an array $X$ is:$~~\sum_0^a X[a]$

- The `cumsum()` function is available through `numpy` 

```python
import numpy as np
cumSumRandomTrials = np.cumsum(randomTrials)

    
print("-" * 30)
print(randomTrials[0:10])
print("-" * 30)
print(list(cumSumRandomTrials[0:10]))
print("-" * 30)
```

- Let's normalize `cumSumRandomTrials` by the number of trials in each position

```python
cumulativePropHeads = cumSumRandomTrials  / range(1, len(cumSumRandomTrials)+1)

print("=====================")
print(randomTrials[0:10])
print("---------------------")
print(cumSumRandomTrials[0:10])
print("---------------------")
print(cumulativePropHeads[0:10])
print("=====================")

```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(12, 8))
plt.plot(range(100), cumulativePropHeads)
plt.axhline(0.5, color='k', linestyle= '--', alpha=0.3)
```

- Notice that even after 1000 draws, the proportion of heads may or may not be equal 0.5. After all, a run of 1000 samples is still just a random sample, and there is no guarantee that the relative frequency in the random sample will match the underlying probability of the event. However, in the "much longer run," the frequency will tend to approximate $p(H)$ and $p(T)$

### Probability versus Statistic

- Probability deals with predicting the likelihood of future events, while statistics analyzes the frequency of past events

- Probability is theoretical branch of mathematics on the consequences of definitions, while statistics is applied  mathematics trying to make sense of real-world observations
